```python
#step activation function -> switch is on or off, har model na alag alag hoi
# vadhare class hoi to soft max, ocha hoi to step 
#bias ne input 1 hamesha
#error current epoch ma zero thay tyare stop thavanu
#bias 0,0, can be random value from â‚¹-1 to 1
#greater than 2 -> soft max, probablity sum 1 thavo joi
#1 vs all probability karvani-> tran hoi to amathi ni max levanu soft max function thi
#output layer ma jetli category atla class
#number of input layer = number of feature
#output layer ma jetla class hoi atla neuron

#jetli column atla input, jetla class atla output neuron

```

```python
import numpy as np
```

```python
#generate 2d and gate using array

x_and = np.array([
            [0,0],
            [0,1],
            [1,0],
            [1,1]
        ])
```

```python
x_and.shape
```

```python
y_and = np.array([0,0,0,1])
```

```python
#weights = np.array([0.3, 0.3]). #model na hisab thi hoi
#bias = 0.0 # aa pan 
#lr = 0.1 # learning chalu thay e pela set karvama aave
#epochs = 1 #learning control thay and throughout learning fix re -> fix re a 50-50 che

#hyper parameter -> during learning constant re, nakki na hoi k a fix re, 50-50 atle kem k tune pan thay
```

```python
weights = np.array([0.5, 0.7]) #model na hisab thi hoi
bias = 0.9 # aa pan 
lr = 0.1 # learning chalu thay e pela set karvama aave
epochs = 5 #learning control thay and throughout learning fix re -> fix re a 50-50 che
```

```python
def step_function(x_and):
    if x_and >= 0:
        return 1
    else:
        return 0
```

```python
for epoch in range(epochs):
    print(f"\nEpoch-> {epoch+1}") #number of epochs print karya
    
    total_mistakes = 0 #total mistakes count karva mate variable
    
    for i in range(4):
        x = x_and[i] #input
        y = y_and[i] #output
        
        z = np.dot(weights, x) + bias #linear function
        y_pred = step_function(z) #activation function
        #w1*x1 + w2*x2 + bias -> z
        print(f"X: {x}, y: {y}, Pred: {y_pred}, z: {z}") #input, predicted output and actual output print karya
        
        
        error = y - y_pred #error calculate kari
        
        #multiple epoch calculate kare say 5 hoi and second ma sachu aavi jai pan aagad na epoch calculate thay a sache j aave because weight update thase j nai
        if error != 0:
            weights = weights + (lr * error * x) #weights update kari, 2D array che to banne na dimesion 
                                                    #same hase to automatic jate thay jase
                                                    #x pote 2d che atle banne input mate thse
            bias = bias + (lr * error * 1) #bias update kari
            total_mistakes += 1 #1 mistake count kari lai chi, early stopping mate
        else:
            print(f"---------------CORRECT {x}----------------") #update ni jarurat nathi to print kari
        
```

```python
def inference_and_gate(a,b):
    #z = np.dot(weights, x) + bias
    z1 = (a*weights[0] + b*weights[1]) + bias 
    pred = step_function(z1)
    return pred
```

```python
inference_and_gate(0,0)
```

```python
inference_and_gate(0,1)
```

```python
inference_and_gate(1,0)
```

```python
inference_and_gate(1,1)
```
