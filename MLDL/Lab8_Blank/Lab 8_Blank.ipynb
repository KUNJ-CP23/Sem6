```python
#select the website link
req = requests.get("https://darshan.ac.in/faculty-list/btech-computer")
```

```python
req.status_code
```

```python
soup = bs4.BeautifulSoup(req.text, 'html.parser')
#soup = bs4.BeautifulSoup(req.text, 'lxml')
```

```python
#for selecting specific data from the webpage
#inspect element and copy css selector
data = soup.select("body > main > div:nth-child(5) > div > div > div > div > div")
```

```python
len(data)
```

```python
#[0] che e array mate text
#span is for selecting span tags jetla hoi atla
for f in data:
    name = f.select("h2")[0].text.strip()
    d = f.select("span")[1].text.strip()
    print(name,"---->",d)
```

```python
for f in data:
    name = f.select("h2")[0].text.strip()
    d = f.select("span")[2].text.strip()
    print(name,"---->",d)
```

```python
#make dataframe 
#converted data into dataframe for better representation
faculty_list = []
for f in data:
    name = f.select("h2")[0].text.strip()
    designation = f.select("span")[1].text.strip()
    email = f.select("span")[2].text.strip()
    faculty_list.append({"Name":name, "Designation":designation, "Email":email})
df_faculty = pd.DataFrame(faculty_list)
```
