## Step 4 . Splitting dataset in to input and output

```python
X = df.drop('Profit', axis=1)
```

```python
y=df[['Profit']]
```

```python
y
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Perform Transformation

```python
# StandardScalar
from sklearn.preprocessing import StandardScaler
num_cols = ['R&D Spend', 'Administration', 'Marketing Spend']

sclar = StandardScaler()
# fit_transform()

X_train_scaled = sclar.fit_transform(X[num_cols])

print("Scaled Data Sample: ")
X_train_scaled
```

```python
x1=pd.get_dummies(X, columns=['State'],drop_first=True)
```

```python
x1
```

## Step 6 . Dummy variable trap

```python
# Already Performed using | drop_first =True
```

```python
plt.scatter(X['R&D Spend'],y)
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.title('R&D Spend vs Profit')
plt.show()
```

## Step 7 Splitting dataset in to Train and Test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x1,y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

## Step 8  Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

## Step 9  Fit the data

```python
model.fit(X_train, y_train)
```

## Step 10  Predict the data

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

## Now use Polynomial Regression on Position_Salaries dataset

```python
df1 = pd.read_csv('Position_Salaries.csv')
```

```python
df1
```

```python
plt.scatter(df1['Level'],df1['Salary'])
plt.show()
```

```python
#feature - target separation
X1 = df1[['Level']]      # feature 2d
y1 = df1['Salary']       # target 1d 
```

```python
#linear regression ma fit karaya
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X1, y1)
```

```python
plt.scatter(X1, y1)
plt.plot(X1, lin_reg.predict(X1))
plt.title("Linear Regression")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
#upar scatter plot ma underfitting che
```

```python
#polynomial feature creation
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X1)
```

```python
#model fitting
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y1)
```

```python
#polynomial regression curse 

X_grid = np.arange(X1.min().values[0], X1.max().values[0], 0.1)
X_grid = X_grid.reshape(-1, 1)

plt.scatter(X1, y1)
plt.plot(X_grid, poly_reg.predict(poly.transform(X_grid)))
plt.title("Polynomial Regression (Degree 4)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
#testing
#prediction
#Predict salary for Level = 6.5

poly_reg.predict(poly.transform([[6.5]]))
```
