## Simple Linear Regression

## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

## Step 2. Import the dataset

```python
df = pd.read_csv('50_Startups.csv')
df
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['State'].nunique()
```

## Step 3 . Check the State Column

```python
print("New York ", (df['State']=='New York').sum())
print("California ", (df['State']=='California').sum())
print("Florida ", (df['State']=='Florida').sum())
```

## Step 4 . Splitting dataset in to input and output

```python
X = df.drop('Profit', axis=1)
```

```python
y=df[['Profit']]
```

```python
y
```

```python
X
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Perform Transformation

```python
# StandardScalar
from sklearn.preprocessing import StandardScaler
num_cols = ['R&D Spend', 'Administration', 'Marketing Spend']

sclar = StandardScaler()
# fit_transform()

X_train_scaled = sclar.fit_transform(X[num_cols])

print("Scaled Data Sample: ")
X_train_scaled
```

```python
x1=pd.get_dummies(X, columns=['State'],drop_first=True)
```

```python
x1
```

## Step 6 . Dummy variable trap

```python
# Already Performed using | drop_first =True
```

```python
plt.scatter(X['R&D Spend'],y)
plt.xlabel('R&D Spend')
plt.ylabel('Profit')
plt.title('R&D Spend vs Profit')
plt.show()
```

## Step 7 Splitting dataset in to Train and Test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x1,y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

## Step 8  Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

## Step 9  Fit the data

```python
model.fit(X_train, y_train)
```

## Step 10  Predict the data

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

## Step 11  Display Result

```python
# y_test and y_predict
```

```python
plt.scatter(X_test['Administration'],y_test)
plt.scatter(X_test['Administration'],y_predict)
```

## RSS

```python
np.sum((y_test.values - y_predict)**2)
```

```python
mean_squared_error = np.sum((y_test.values - y_predict)**2)/len(y_test)
```

## R Square

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_predict)
```
